```
from bateaux_metrics_v2 import *
```

```
parent_dir = os.path.dirname(os.getcwd())
model_dir = os.path.join(parent_dir, 'models', 'Attention')
model_name = 'att_model_v4.hdf5'
model_path = os.path.join(model_dir, model_name)
```

```
custom_objects = {'loss': euclidean_loss(beta), 
                   'euclidean_dist':euclidean_dist, 
                   'x_error':x_error, 
                   'y_error':y_error, 
                   'z_error':z_error}
```

```
model = load_model(model_path, custom_objects=custom_objects)
```

```
test_file = os.path.join(parent_dir, 'sets_labels', 'combined', 'shuffled', 'test_set.hdf5')
h5_reader = h5py.File(test_file, 'r')
list(h5_reader.keys())
```

```
x_test = np.array(h5_reader['test'])
h5_reader.close()
```

```
# since images are single channel
x_test = x_test.astype('float32')
x_test /= 255
n, l, w = x_test[0], x_test[1], x_test[2]
x_test = np.stack((x_test, x_test, x_test), axis=-1)
print(x_test.shape)
```

```
test_labels = os.path.join(parent_dir, 'sets_labels',  'combined', 'shuffled', 'test_labels.csv')
y_test = np.loadtxt(test_labels, dtype=np.float32)
```

```
metric_names = model.metrics_names
scores = model.evaluate(x_test, y_test, verbose=1)
for metric, val in zip(metric_names, scores):
  print("{} : {}".format(metric, val))
```
